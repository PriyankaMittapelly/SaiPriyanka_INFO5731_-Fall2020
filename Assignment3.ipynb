```python
import numpy as np
import matplotlib.pyplot as plt
```

```python


records=[10,500,1000,5000,10000,50000,80000,100000,200000]

list_pi=[]

for lis in records:
    
  #generate list of random number for both x, y coordinates
    x=np.random.uniform(0,1,[lis,1])
    y=np.random.uniform(0,1,[lis,1])

  #check to see if point is inside or outside circle
    inside_circle = ((x-0.5)**2+(y-0.5)**2) <0.25


  # calaculate pi value
    pi=4*np.sum(inside_circle)/lis
    list_pi.append(pi)


  #filter for list of points inside the circle
    x_in=x[inside_circle]
    y_in=y[inside_circle]

```

```python
#ploting the graph

fig=plt.figure()
plt.scatter(x, y, c='b', s=1)
plt.scatter(x_in, y_in, color='r',s=1)
circle1 = plt.Circle((0.5,0.5), 0.5, color='black',fill=False)
plt.gcf().gca().add_artist(circle1)
plt.show()
print("value of pi for "+str(lis)+" records is"+str(pi))
```
