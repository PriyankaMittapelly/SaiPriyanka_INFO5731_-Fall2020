```python
# Question 2

File.to_csv("File_output.csv")

File.columns
```

```python
data=File.drop(columns=File.iloc[:,1:22])
data["mean"]=data.iloc[:, 1:-1].mean(axis=1)
data.sort_values(by="continent",ascending=True,inplace=True)
data=data.drop(columns=data.iloc[:,1:-2])
data
```

```python
df=pd.DataFrame(data.groupby(["continent","Estimated HIV Prevalence% - (Ages 15-49)"])["mean"].max())
Avg = df.groupby('continent')['mean'].idxmax()
Avg=df.loc[Avg]
Avg=Avg.reset_index()
Avg

```

```python
df=pd.DataFrame(data.groupby(["continent","Estimated HIV Prevalence% - (Ages 15-49)"])["mean"].min())
lavg=df.groupby('continent')['mean'].idxmin()
lavg=df.loc[lavg]
lavg=lavg.reset_index()
lavg
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(rc={'figure.figsize':(9.9,9.29)})
sns.set(style="whitegrid")
plt.title("Highest average of continent and country",size=25)
sns.barplot(x="mean",y="Estimated HIV Prevalence% - (Ages 15-49)",hue="continent",data=Avg)
plt.show()
plt.title("lowest average of continent and country",size=25)
sns.barplot(x="mean",y="Estimated HIV Prevalence% - (Ages 15-49)",hue="continent",data=Avg)
plt.show()
```

```python
continent=list(Avg["continent"])
plt.bar(continent, Avg["mean"] ,label=Avg[["continent","Estimated HIV Prevalence% - (Ages 15-49)","mean"]],color="BLUE")
plt.bar(continent,lavg["mean"],width=0.50,label=lavg[["continent","Estimated HIV Prevalence% - (Ages 15-49)","mean"]],color="RED")
plt.xticks(color='BLACK', rotation=40)
plt.legend()
```

```python
sns.set(rc={'figure.figsize':(9,5)})
sns.lineplot(continent,Avg["mean"],color="blue")
sns.lineplot(continent,lavg["mean"],color="yellow")
```
